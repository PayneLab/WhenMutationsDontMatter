```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats
import gseapy as gp
import re
import sys 

import cptac
import cptac.utils as u
import plot_utils as p
```

```python
luad_df = pd.read_csv("csv/Single_Cancer/Luad_phospho_MAPK.csv")
luad_sig = luad_df.loc[luad_df['Luad_P_Value'] < 0.1]
luad_sig
```

```python
luad_sig['Protein'], luad_sig['Site'] = luad_sig['Phospho'].str.split('_', 1).str
```

```python

luad_sig['mean_Median'] = luad_sig.groupby('Protein')['Luad_Median'].transform('mean')

luad_sig['mean_P_Value'] = luad_sig.groupby('Protein')['Luad_P_Value'].transform('mean')

cytoscape_table = luad_sig[["Protein","mean_Median", "mean_P_Value"]]
cytoscape_table.drop_duplicates(inplace=True)
cytoscape_table = cytoscape_table.set_index("Protein")
```

```python
luad_sig
```

```python
cytoscape_table.to_csv("Luad_phospho_avg.csv", index = True)
```

```python
Endo_df = pd.read_csv("csv/Single_Cancer/Endo_phospho_MAPK.csv")
Endo_sig = Endo_df.loc[Endo_df['Endo_P_Value'] < 0.1]
Endo_sig['Protein'], Endo_sig['Site'] = Endo_sig['Phospho'].str.split('_', 1).str
Endo_sig
```

```python

Endo_sig['mean_Median_Endo'] = Endo_sig.groupby('Protein')['Endo_Median'].transform('mean')

Endo_sig['mean_P_Value_Endo'] = Endo_sig.groupby('Protein')['Endo_P_Value'].transform('mean')

cytoscape_table = Endo_sig[["Protein","mean_Median_Endo", "mean_P_Value_Endo"]]
cytoscape_table.drop_duplicates(inplace=True)
cytoscape_table = cytoscape_table.set_index("Protein")
```

```python
cytoscape_table
```

```python
cytoscape_table.to_csv("Endo_phospho_avg.csv", index = True)
```
