Get tables with MAPK signaling phosphoproteomics.

```python
df = pd.read_csv("../Make_Tables/csv/phospho_MAPK.csv")
luad_df = pd.read_csv("../Make_Tables/csv/Single_Cancer/LUAD_phospho_MAPK.csv")
e_df = pd.read_csv("../Make_Tables/csv/Single_Cancer/EC_phospho_MAPK.csv")
```

# Count unique sig phosphosites for each protein

```python
luad_sig = luad_df.loc[luad_df['LUAD_P_Value'] < 0.1]
e_sig = e_df.loc[e_df['EC_P_Value'] < 0.1]
```

```python
# proteins affected multiple times p < .1
luad_sig['Protein'], luad_sig['Site'] = luad_sig['Phospho'].str.split('_', 1).str
luad = luad_sig.Protein.value_counts().to_frame('LUAD_Sig_Phospho')

e_sig['Protein'], e_sig['Site'] = e_sig['Phospho'].str.split('_', 1).str
ec = e_sig.Protein.value_counts().to_frame('EC_Sig_Phospho')

m = luad.join(ec, how = 'outer').sort_values(by = ['LUAD_Sig_Phospho', 'EC_Sig_Phospho'], ascending = False)
m
```

```python
# Create csv file
m.to_csv('Sup_Table_2.csv')
```
