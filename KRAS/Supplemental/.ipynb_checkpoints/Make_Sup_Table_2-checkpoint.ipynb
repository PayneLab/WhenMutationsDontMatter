# Table of number of sites affected for trans proteins

```python
import pandas as pd
import numpy as np

import cptac
import cptac.utils as u
import plot_utils as p

import warnings
warnings.filterwarnings('ignore')
```

Get tables with MAPK signaling phosphoproteomics.

```python
df = pd.read_csv("../Make_Tables/csv/phospho_MAPK.csv")
luad_df = pd.read_csv("../Make_Tables/csv/Single_Cancer/Luad_phospho_MAPK.csv")
e_df = pd.read_csv("../Make_Tables/csv/Single_Cancer/Endo_phospho_MAPK.csv")
```

# Get counts of sites affected for each protein

```python
luad_sig = luad_df.loc[luad_df['Luad_P_Value'] < 0.1]
e_sig = e_df.loc[e_df['Endo_P_Value'] < 0.1]
```

```python
# proteins affected multiple times p < .1
luad_sig['Protein'], luad_sig['Site'] = luad_sig['Phospho'].str.split('_', 1).str
luad = luad_sig.Protein.value_counts().to_frame('Luad_Sig_Phospho')

e_sig['Protein'], e_sig['Site'] = e_sig['Phospho'].str.split('_', 1).str
ec = e_sig.Protein.value_counts().to_frame('Endo_Sig_Phospho')

m = luad.join(ec, how = 'outer').sort_values(by = ['Luad_Sig_Phospho', 'Endo_Sig_Phospho'], ascending = False)
m
```

```python
m.to_csv('Supplemental_Table_2_KRAS.csv')
```
