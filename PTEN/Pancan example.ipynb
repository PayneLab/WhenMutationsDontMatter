Here is a list of colleges that provided different data sources:
* School of Medicine at Mt. Sinai: mssm
* Washington University: washu 
* University of Michigan: umich
* Baylor College of Medicine: bcm
* Broad Institute: broad

The Proteomic Data Commons (pdc) also processed data. https://pdc.cancer.gov/pdc/

```python
g.list_data_sources() 
```

The clinical file from Mt. Sinai (mssm) contains the master index of all the patient IDs. For example, some IDs are included in proteomics but not in transcriptomics. All IDs found in any omics df are included in the clinical df from mssm. 

```python
clin = g.get_clinical('mssm')
clin
```

We also grouped information from the large mssm clinical file to make the demographic, medical conditions, previous cancer, cancer diagnosis, and followup dfs. 

```python
mc = g.get_medical_conditions('mssm')
mc
```

```python
prot = g.get_proteomics('pdc')
prot
```
