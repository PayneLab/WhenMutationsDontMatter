# Create Heatmaps for significant Immune genes

Pancancer heatmaps are created with circle size showing significance and color showing differences in median.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats
import gseapy as gp
import re
import sys 

import cptac
import cptac.utils as u

import plot_utils as p
```

# Sig in multiple cancers

```python
root = R'~\Github\WhenMutationsDontMatter\PTEN\Step_3_trans_effect\csv'
mult_sig_df = pd.read_csv(root+R'\mult_sig_pval_heatmap.csv')
```

```python
tcr = ['CHUK', 'PRKCB', 'WAS', 'PTPRC', 'CD4', 'ZAP70', 'MAP4K1']
bcr = ['CYTH1', 'SYK', 'BTK', 'PLCG2', 'PIK3CD', 'BLNK', 'PIK3CA',
       'ZAP70']
cxc = ['ELMO1', 'DOCK2', 'PRKCB', 'GNAI2', 'GNG2', 'RAC2']
immune = [tcr, bcr, cxc]

combined = []
for pathway_list in immune:
    for gene in pathway_list:
        combined.append(gene)

print('total genes:',len(combined))
```

```python
bool_df = mult_sig_df.Proteomics.isin(combined)
plot_df2 = mult_sig_df[bool_df]
plot_df2.Proteomics.unique()
```

```python
# Only include p-values < 0.05
plot_df2 = plot_df2.loc[plot_df2['P_Value'] <= 0.05]
```

```python
p.plotCircleHeatMap(plot_df2, circle_var = 'P_Value', color_var='Medians', x_axis= 'Proteomics', y_axis = 'Cancer',
                    plot_height=200, plot_width=500, legend_min= 1e-3, legend_max=0.05)
```
