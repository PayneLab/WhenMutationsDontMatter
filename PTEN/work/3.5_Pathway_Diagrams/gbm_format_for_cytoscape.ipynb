```python
root = R'~\Github\WhenMutationsDontMatter\PTEN\Step_3_trans_effect\csv'
gbm = pd.read_csv(root+R'\gbm_pval_medians.csv')
gbm = gbm.replace('_proteomics','',regex=True)
```

```python
def replace(items, a, b): 
    return [b if x == a else x for x in items]

```

```python
# covert gene name to equivilent name in endo dataset
dna_rep = u.get_proteins_in_pathway('DNA Replication')
dna_rep = replace(dna_rep,'ORC1L','ORC1')
dna_rep = replace(dna_rep,'ORC2L','ORC2')
dna_rep = replace(dna_rep,'ORC3L','ORC3')
dna_rep = replace(dna_rep,'ORC4L','ORC4')
dna_rep = replace(dna_rep,'ORC5L','ORC5')
dna_rep = replace(dna_rep,'ORC6L','ORC6')
dna_rep = replace(dna_rep,'POLA','POLA1')
dna_rep = replace(dna_rep,'PRIM2A','PRIM2')

dna_rep
```

```python
b = gbm['Proteomics'].isin(dna_rep)
dna_rep_df = gbm[b]
dna_rep_df
```

```python
# covert gene names to equivilent names to match shared_name in cytoscape
orc_genes = ['ORC1','ORC2','ORC3','ORC4','ORC5','ORC6']
for orc in orc_genes:
    dna_rep_df = dna_rep_df.replace(orc, orc+'L', regex=True )
dna_rep_df = dna_rep_df.replace('POLA1', 'POLA', regex=True )
dna_rep_df
```

```python
dna_rep_df.to_csv(root+R'\gbm_dna_rep.csv',index=False)
```
