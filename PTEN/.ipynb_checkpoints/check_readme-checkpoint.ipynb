```python
import pandas as pd
import numpy as np
import warnings
from functools import reduce
import re
import cptac
import cptac.pancan as pc
```

```python
g = pc.PancanGbm()
l = pc.PancanLuad()
ls = pc.PancanLscc()
e = pc.PancanUcec()
r = pc.PancanCcrcc()
h = pc.PancanHnscc()
b = pc.PancanBrca()
o = pc.PancanOv()
c = pc.PancanCoad()
pdc = pc.PancanPdac()
```

```python
cas = {'GBM':g, 'HNSCC':h, 'LSCC':ls, 'LUAD':l, 'BRCA':b, 'OV':o, 'UCEC':e, 'COAD':c, 'CCRCC':r, 'PDAC':pdc}
```

```python
source = 'pdc'
for cancer in cas:
    print('\n\n')
    print(cancer)
    ca = cas[cancer]
    
    cl = ca.get_clinical(source)
    print('clin:')
    #print(set(cl.index.duplicated()))
    #print(cl.iloc[:,:5].head())
    print(cl.index.to_list())
    
    p = ca.get_proteomics(source)
    ph = ca.get_phosphoproteomics(source)
    print('prot:')
    #print(set(p.index.duplicated()))
    #print(p.iloc[:,:5].head())
    print(p.index.to_list())
    
    
    print('\nphospho:')
    #print(ph.iloc[:,:5].head())
    #print(set(p.index.duplicated()))
    print(ph.index.to_list())
    
    
    try:
        print('\nacetyl:')
        a=ca.get_acetylproteomics('pdc')
        #print(a.iloc[:,:5].head())
        #print(set(a.index.duplicated()))
        print(a.index.to_list())
    except:
        print('no acetyl')
        continue
    
```
