# Filter out hotspot mutations
Hotspot:
E542K
E545K
H1047R

## Questions: 
When I filter for hotspots above, I end up with only five samples.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats
import gseapy as gp
import re
import sys 

import cptac
import cptac.utils as u
```

```python
gene = "PIK3CA"
```

```python
en = cptac.Endometrial()
br = cptac.Brca()
cl = cptac.Colon()
```

# PROTEOMICS

## Endometrial

```python
mut = en.get_genotype_all_vars(gene, mutation_hotspot=['E542K', 'E545K', 'H1047R'])
prot = en.get_proteomics()
protfilter = prot.index.str.endswith('.N')
prot = prot[~protfilter]

joined = mut.join(prot)
joined
```

```python
joined[joined.Location.str.contains('E542K')]
```

```python
joined[joined.Location.str.contains('E545K')]
```

```python
joined[joined.Location.str.contains('H1047R')]
```
