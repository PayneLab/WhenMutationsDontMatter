```python
#deal with multiindex

#filter out the normal 
filtered = en.reduce_multiindex(df=omics_mutations, levels_to_drop="Site").head()
mutation = filtered.loc[filtered['Sample_Status'] == "Tumor"]
mutations = mutation.drop(columns=[gene+"_Mutation", gene+"_Location", gene+"_Mutation_Status", "Sample_Status"])
mutations.head()

#do for the first site
mutations = mutations.iloc[:,0:1]
mutations['PIK3CA_phosphoproteomics'].unique()
```

```python
mutation_types_df = df.get_genotype_all_vars(gene)
mutation_types_df['Mutation'] = [','.join(map(str, l)) for l in mutation_types_df['Mutation']]
plotting_df = pd.concat([mutation_types_df, mutations], axis=1)
    
#sort the df so that the same mutations will show up for every tumor type
plotting_df = plotting_df.sort_values('Mutation')

# plotting_df.head()
```

```python
a4_dims = (11.7, 8.27) #dimensions for bigger plot
fig, ax = plt.subplots(figsize=a4_dims)

somatic_boxplot = sns.boxplot(data=plotting_df, x="Mutation",y=gene +"_phosphoproteomics", ax = ax) 
                                        # order parameter is used to reorder the mutation categories 

somatic_boxplot.set_title("PIK3CA gene mutation protein abundance")
somatic_boxplot = sns.stripplot(data=plotting_df, x="Mutation",y=gene +"_phosphoproteomics",jitter=True, 
                                color=".3")
somatic_boxplot.set(xlabel="Somatic Gene Mutation",ylabel="phosphoproteomics")
somatic_boxplot.set_xticklabels(somatic_boxplot.get_xticklabels(),rotation=30) # Rotate labels to avoid overlap
plt.show()

# omics_mutations.columns.get_level_values("Name")

# gene1_filter = omics_mutations.columns.get_level_values("Name").str.contains("Sample_Status") # Select all columns where the gene starts with "AA". This will grab every column where the key "Name" starts with AA
# gene1_data = omics_mutations.loc[:, gene1_filter]
# gene1_data.head()
```
