```python
def HasPosNeg(row):
    hasPos = False

    for item in row:
        if pd.isnull(item):
            continue
        if item > 0.3:
            hasPos = True
            
    if hasPos:
        return True
    return False
```

```python
df_GBM = pd.read_csv("Step3.1_Pearson_dfs_by_cancer/csv_files/GBM_EGFR_all_pearson_FDR.csv")
df_GBM = df_GBM.drop(['Unnamed: 0'], axis=1)
df_GBM = df_GBM.set_index('Comparison')
df_GBM
```

```python
df_GBM_pos =df_GBM[df_GBM.Correlation > 0]
df_GBM_pos
```

```python

df1_transposed = df_GBM_pos.T 
```

```python

FDR_pos = df1_transposed.columns.values.tolist()
pos_GBM_genes = []
for gene in FDR_pos:
    pos_GBM_genes.append((re.sub("_proteomics", "", gene)))
len(pos_GBM_genes)
```

```python

tumor_enr = gp.enrichr(gene_list = pos_GBM_genes, description='Tumor_partition', gene_sets='KEGG_2016')
```
