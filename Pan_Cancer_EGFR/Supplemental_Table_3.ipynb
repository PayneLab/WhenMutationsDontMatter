# Supplemental Table 3

Manuscript references 4,149/ 6,230 trans genes are significant in only 1 cancers. This notebook derives that number and also includes the break down of how many trans genes are significant in 1,2,3 ect cancers. 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats
import re
import sys 
import statsmodels.stats.multitest

import cptac
```

```python
def Pvalue_sig(row):
    numSig = 0

    for item in row:
        if pd.isnull(item):
            continue
        if item < 0.05:
            numSig += 1
            
    return numSig
```

```python
#Load in table created for figure 1C. (Includes all correlations and FDR corrected p values for all proteins)
df_fdr_pvals = pd.read_csv("../Make_EGFR_Fig_1C/Pval_corr_table_Fig_1C.csv")
df_fdr_pvals = df_fdr_pvals.drop(['Unnamed: 0'], axis=1)
df_fdr_pvals = df_fdr_pvals.set_index("Comparison")
df_fdr_pvals = df_fdr_pvals[1:] #Drop EGFR 
```

```python
df_fdr_pvals["Num_sig"] = df_fdr_pvals.loc[:,["P_value_Gbm","P_value_kidney","P_value_Ovar","P_value_Brca","P_value_Luad","P_value_Hnscc","P_value_Lscc","P_value_Colon"]].apply(Pvalue_sig, axis = 1)
df_fdr_pvals = df_fdr_pvals.sort_values(by=['Num_sig'],ascending=False)
```

```python

df_fdr_pvals_1 = df_fdr_pvals[df_fdr_pvals['Num_sig'].isin([1,2,3,4,5,6,7,8])]
df_fdr_pvals_1
```

```python
counts = df_fdr_pvals_1['Num_sig'].value_counts()
counts
```
